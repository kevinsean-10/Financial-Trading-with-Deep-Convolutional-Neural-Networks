# Data Modelling  
## Splitting and Scaling Data

In this stage, we prepare the data for modeling by splitting it into training and validation sets and applying feature scaling.

We set the spatial dimension (`dim`) to **18**, which allows each sample's feature vector to be reshaped into an **18×18 matrix**. This transformation is crucial for enabling the use of convolutional layers in the model architecture.

````python
selected_columns = list(cleaned_data.drop(columns=['Date', 'Labels']).columns)
feature_columns = list(cleaned_data.drop(columns=['Date', 'Labels', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']).columns)
# feature_columns_idx = range(6, 202)
feature_columns_idx = [0,3] + list(range(6, 328))

dim = 18
num_features = int(dim**2)

preprocess.splitting_and_scaling(train_size=0.8, validation_size=0.2, selected_columns=selected_columns, print_output=True)
````

## Feature Selection

````python
preprocess.feature_selection(feature_columns_idx=feature_columns_idx)
preprocess.data_figuring(dim=dim)
````

## Creating and Executing Model Architecture

After experimenting with various model configurations and performing hyperparameter tuning, we determined that the following architecture yields the best performance for our dataset.

The selected model structure is optimized to effectively capture the patterns in the engineered features derived from technical indicators.

````python
params = {'batch_size': 64, 
          'conv2d_layers': {'conv2d_do_1': 0.2, 'conv2d_filters_1': 30, 'conv2d_kernel_size_1': 2, 'conv2d_mp_1': 0,
                            'conv2d_strides_1': 1, 'kernel_regularizer_1': 0.0, 
                            'conv2d_do_2': 0.15,'conv2d_filters_2': 20, 'conv2d_kernel_size_2': 2, 'conv2d_mp_2': 2, 
                            'conv2d_strides_2': 2, 'kernel_regularizer_2': 0.0, 
                            'layers': 'two'
                           },
          
          'dense_layers': {'dense_do_1': 0.2, 
                           'dense_nodes_1': 100, 
                           'kernel_regularizer_1': 0.0, 
                           'dense_do_2': 0.2,
                           'dense_nodes_2': 16,
                           'layers': 'one'
                          },
          'epochs': 400, 
          'lr': 0.002,
          'optimizer': 'adam',
         }

preprocess.modeling(params=params, print_summary=True, plot_history=True)
preprocess.evaluate_model()
````
